### Step 1
- Import pandas as pd
- Then import matplotlib.pyplot as plt
- Then set the %matplotlib inline
- Set the figure.figsize parameter to (5,2) so that it fits nicely on the screen.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.figure(figsize= (5,2))
```

### Step 2
- Read the data from CSV file stock-data.csv
- Use the index_col=0
- Also, it is convenient to use parse_dates=True
- Take head of data

```python
read_csv = pd.read_csv('stock_data.csv', index_col = 0, parse_dates= True)
```

```python
read_csv.head()
```

### Step 3
- make a plot of Close price.

```python
plt.plot(read_csv['Close'])
plt.xlabel('Date')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### Step 5
- Make a figure and axes by calling matplotlib.pyplot from plt
- Now make a chart with Closing price

```python
plt.plot(read_csv['Close'])
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Step 6
- Make new figure of 4 subplots in 2 rows and 2 columns
- Make charts of Open, High, Low, and Close in each axes

```python
# Plot 1:
fig, axs = plt.subplots(2, 2, figsize=(10.5, 8))

ax1 = axs[0, 0]
ax1.plot(read_csv['Open'])
ax1.set_xlabel('Date')

ax3 = axs[1, 0]
ax3.set_xlabel('Date')
ax3.plot(read_csv['High'])

ax2 = axs[0, 1]
ax2.plot(read_csv['Low'])
ax2.set_xlabel('Date')

ax4 = axs[1, 1]
ax4.plot(read_csv['Close'])
ax4.set_xlabel('Date')

ax1.set_xticklabels(read_csv.index.strftime('%Y-%m'), rotation=45)

ax2.set_xticklabels(read_csv.index.strftime('%Y-%m'), rotation=45)

ax3.set_xticklabels(read_csv.index.strftime('%Y-%m'), rotation=45)

ax4.set_xticklabels(read_csv.index.strftime('%Y-%m'), rotation=45)

plt.tight_layout()
plt.show()
```

### Step 7
- Make a new figure
- Make a bar chart with volume for January 2020

```python
fig = plt.figure(figsize = (10, 5))
 
read_csv.index.values

 # Filter dates with January month
january_dates = read_csv[(read_csv.index.month == 1 ) & (read_csv.index.year == 2020 )]

# Append '00:00:00' to the timestamp of the filtered dates
january_dates.index = january_dates.index.strftime('%Y-%m-%d') + ' 00:00:00'

january_dates = january_dates.sort_index()

plt.bar(january_dates.index , january_dates['Volume'], color ='steelblue',
        width = 0.5)

plt.xlabel("Date")
plt.xticks(rotation=90)
plt.show()
```
